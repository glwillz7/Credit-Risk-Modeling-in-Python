# PD Model

# Data Preparation

 EL = PD * LGD * EAD. Established methodology to model probability of default is a logistic regression where dependent variable is precisely whether a customer default or not.  Logarithm of odds of an outcome of interest and a linera combination of predictors

Deafult: 0, Non-default: 1 Many definitions of default: 90 days overdue payment, committed a fraud.

PD model: Logisitic regression (LR). Dependent variable (output variable) ranges between 0 (default) and 1 (non-default).

LR estimates the relationship between ln(odds) of outcome variable and a linear combination of independent variables. ln(odds) = ln(non-defaults / defaults)

PD model should be easy to use and understand. So, all independent variables in PD model should be dummy variables. Discrete variables have already been converted into dummy variables. Continuous variables too should be converted.

# Automating Calculations

Perform calculations for any given pair of categorical(binary) and dep variables

Define functions with def statement. 
df: independent variables data frame df
discrete_variable_name = discrete local dependent variable,
good_bad_variable_df = outcome 

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
               df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1- df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good']/ df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad']/ df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good']/ df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
    
```

# Preprocessing Continuous Variables: Automating calculations and visualizing results

# Term Variable

The same function used for discrete variables can be used for continuous variables. Discrete variables don't have qualitative comparison (so, sorted them by WOE). 
But continuous variables have a quantitative comparison (so,leave them in natural order). 
For example, all values in 90k are less than 100k

```python
def woe_ordered_continuous(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
               df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1- df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good']/ df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad']/ df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good']/ df['prop_n_bad'])
    #df = df.sort_values(['WoE'])
    #df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
